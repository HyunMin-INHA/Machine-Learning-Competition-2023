```python
## Now we are going to visualize Histogram of GR in training data #!

plt.figure(figsize=(3,3))
plt.hist(df.GR.values, bins=20)
plt.xlabel('GR')
plt.ylabel('Frequency')
plt.title('Gamma-ray record in Train #1')
plt.grid('on')

## plt.savefig('GR_HIST_TRAIN_1.png')
```

```python
fig, ax = plt.subplots(figsize=(3,10))

plt.plot(df['GR'], df['DEPT'])
plt.gca().invert_yaxis() ## y축 플립하는 방법
plt.xlabel('GR')
plt.ylabel('DEPTH')
plt.grid('on')
```

```python
plt.figure(figsize=(3,10))
plt.scatter(df['GR'], df['DEPT'])
```

```python
plt.figure(figsize=(12, 10))
plt.suptitle('Training data #1')

plt.subplot(1,4,1)
plt.plot(df['GR'], df['DEPT'])
plt.gca().invert_yaxis() ## y축 플립하는 방법
plt.title('Gamma-ray log')
plt.xlabel('Unit : API')
plt.ylabel('DEPTH')
plt.grid('on')

plt.subplot(1,4,2)
plt.plot(df['RHOB'], df['DEPT'])
plt.gca().invert_yaxis() ## y축 플립하는 방법
##plt.gca().axes.yaxis.set_visible(False)
plt.xlim(2,2.75)
plt.title('Density log')
plt.xlabel('Unit : g/cc')
##plt.ylabel('DEPTH')
plt.grid('on')
plt.tight_layout()

plt.subplot(1,4,3)
plt.plot(df['NPHI'], df['DEPT'])
plt.gca().invert_yaxis() ## y축 플립하는 방법
##plt.gca().axes.yaxis.set_visible(False)
plt.title('Neutron log')
plt.xlabel('Unit : %')
##plt.ylabel('DEPTH')
plt.grid('on')

plt.subplot(1,4,4)
plt.plot(df['RD'], df['DEPT'])
plt.gca().invert_yaxis() ## y축 플립하는 방법
##plt.gca().axes.yaxis.set_visible(False)
plt.xlim(0,30)
plt.title('Resistivity log')
plt.xlabel('Unit : Ohm-m')
##plt.ylabel('DEPTH')
plt.grid('on')

plt.tight_layout()
```

```python
for i in range(1, 10):
    df = pd.read_csv('data/train/aligned_well_0{}.csv'.format(i))
    
    plt.figure(figsize=(12, 10))
    plt.suptitle('Training data #{}'.format(i))

    plt.subplot(1,4,1)
    plt.plot(df['GR'], df['DEPT'])
    plt.gca().invert_yaxis() ## y축 플립하는 방법
    plt.title('Gamma-ray log')
    plt.xlabel('Unit : API')
    plt.ylabel('DEPTH')
    plt.grid('on')

    plt.subplot(1,4,2)
    plt.plot(df['RHOB'], df['DEPT'])
    plt.gca().invert_yaxis() ## y축 플립하는 방법
    ##plt.gca().axes.yaxis.set_visible(False)
    plt.xlim(2,2.75)
    plt.title('Density log')
    plt.xlabel('Unit : g/cc')
    ##plt.ylabel('DEPTH')
    plt.grid('on')
    plt.tight_layout()

    plt.subplot(1,4,3)
    plt.plot(df['NPHI'], df['DEPT'])
    plt.gca().invert_yaxis() ## y축 플립하는 방법
    ##plt.gca().axes.yaxis.set_visible(False)
    plt.title('Neutron log')
    plt.xlabel('Unit : %')
    ##plt.ylabel('DEPTH')
    plt.grid('on')

    plt.subplot(1,4,4)
    plt.plot(df['RD'], df['DEPT'])
    plt.gca().invert_yaxis() ## y축 플립하는 방법
    ##plt.gca().axes.yaxis.set_visible(False)
    plt.xlim(0,30)
    plt.title('Resistivity log')
    plt.xlabel('Unit : Ohm-m')
    ##plt.ylabel('DEPTH')
    plt.grid('on')

    plt.tight_layout()
    plt.savefig('Train{}_log.png'.format(i))
```
